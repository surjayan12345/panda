```python
def win_record(team1,team2):
    
    mask1=data['team1']==team1
    mask2=data['team2']==team2
    mask3=data['team1']==team2
    mask4=data['team2']==team1
    modified_data=data[mask1 & mask2]
    modified_data=data[mask3 & mask4]
    result=modified_data['winner'].value_counts()+modified_data['winner'].value_counts()
    return result
team1=str(input("enter the first team:"))
team2=str(input("enter the second team"))
print(win_record(team1,team2))
```
