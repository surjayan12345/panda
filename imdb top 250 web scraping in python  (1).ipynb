```python
url1='https://www.imdb.com/search/title/?groups=top_250&sort=user_rating'
url2='https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt'
url3='https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt'
url4='https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt'
url5='https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=201&ref_=adv_nxt'
urls=[url1,url2,url3,url4,url5]
rank=[0]
movie_name=[]
year=[]
genre=[]
rating=[]
director=[]
metascore=[]
all_star=[]
votes=[]
gross=[]
duration=[]
certificate=[]
for url in urls:
    page=requests.get(url)
    soup =BeautifulSoup(page.text,'html.parser')
    content=soup.find_all(class_='lister-item-content')
    for i in content:
        # Rank
        rank.append(rank[-1]+1)
        
        # MOVIE NAME
        movie_name.append(i.find_all('a')[0].text)

        # YEAR OF RELEASE
        year.append(i.find('span',{'class':'lister-item-year text-muted unbold'}).text)

        # GENRE
        genre.append(i.find('span',{'class':'genre'}).text[1:8])

        # RATING 
        rating.append(i.find('strong').text)

        # METASCORE
        try:
            metascore.append(i.find('div',{'class':'inline-block ratings-metascore'}).find('span').text) 
        except:
            metascore.append('None')

         # DIRECTOR   
        director.append(i.find_all('a')[12].text)       

        # STARS
        star=[i.find_all('a')[j].text for j in range(13,17)]# star
        all_star.append(star)

        #VOTES
        votes.append(i.find('span',attrs={'name':'nv'}).text) 

        # GROSS 
        try:
            gross.append(i.find_all('span',attrs={'name':'nv'})[1].text)
        except:
            gross.append('None data')

        # DURATION 
        duration.append(i.find('span',{'class':'runtime'}).text) 

        # CERTIFICATE
        try:
            certificate.append(i.find('span',{'class':'certificate'}).text)
        except:
            certificate.append('None')
```

```python
final_array=list(zip(rank,movie_name,year,genre,rating,director,metascore,all_star,votes,gross,duration,certificate))
df=pd.DataFrame(final_array,columns=['Rank','Movie_Name','Year_of_Release','Genre','Rating','Director','MetaScore','Stars','Votes','Gross','Duration','Certificate'])
df
```
